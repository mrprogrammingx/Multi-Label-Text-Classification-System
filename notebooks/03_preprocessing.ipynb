```python
import sys
from pathlib import Path

project_root = Path.cwd().parent
sys.path.append(str(project_root / "src"))

import pandas as pd

from member3_preprocess import (
  clean_text,
  preprocess_dataframe,
  show_examples,
  TextPreprocessor
)
```

```python
train_df = pd.read_csv("../data/raw/dataset_C_train.csv")
val_df = pd.read_csv("../data/raw/dataset_C_val.csv")

train_df.head()
```

Check dataset columns and size

```python
print("Train shape:", train_df.shape)
print("Validation shape:", val_df.shape)

print("\nColumns:")
print(train_df.columns.tolist())
```

Show before/after examples

```python
show_examples(train_df, text_column="text", n=5)
```

Apply preprocessing to train and validation

```python
train_clean = preprocess_dataframe(train_df, text_column="text")
val_clean = preprocess_dataframe(val_df, text_column="text")

train_clean.head()
```

Check empty texts after cleaning

```python
print("Empty rows after cleaning:")
print("Train:", train_clean["is_empty_after_clean"].sum())
print("Validation:", val_clean["is_empty_after_clean"].sum())
```

Check how many texts changed

```python
changed_train = (train_df["text"].apply(clean_text) != train_clean["text"]).sum()
changed_val = (val_df["text"].apply(clean_text) != val_clean["text"]).sum()

print("Changed train texts:", changed_train)
print("Changed validation texts:", changed_val)
```
